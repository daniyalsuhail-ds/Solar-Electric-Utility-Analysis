```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data Preparation
initial_investment_without_itc = 53336.29
initial_investment_with_itc = 37335.40
annual_savings = 15224.15

# 1. Annual Solar Savings
energy_prices = [0.087, 0.122, 0.286]  # Off-peak, mid-peak, on-peak
annual_savings_list = [annual_savings * p / 0.165 for p in energy_prices]
states = ["Off-Peak", "Mid-Peak", "On-Peak"]

plt.figure(figsize=(10, 6))
plt.plot(energy_prices, annual_savings_list, marker='o')
plt.title("Annual Solar Savings vs. State Energy Prices")
plt.xlabel("Energy Prices (CAD/kWh)")
plt.ylabel("Annual Solar Savings (CAD)")
plt.xticks(energy_prices, states)
plt.grid(True)
plt.savefig("Annual_Solar_Savings_vs_Energy_Prices.pdf")
plt.show()

# 3. Benefit Cash Inflows
benefit_data = {
    "Category": ["Initial Investment (CAD)", "Annual Operating Cost (CAD)", "Annual Savings (CAD)"],
    "Shore Power": [0, 28440.80, 0],
    "Solar + Shore Power": [initial_investment_without_itc, 13216.65, annual_savings]
}

df_benefit = pd.DataFrame(benefit_data)

fig, ax = plt.subplots(figsize=(10, 6))
index = np.arange(len(df_benefit["Category"]))
bar_width = 0.35

bar1 = plt.bar(index, df_benefit["Shore Power"], bar_width, label='Shore Power')
bar2 = plt.bar(index + bar_width, df_benefit["Solar + Shore Power"], bar_width, label='Solar + Shore Power')

plt.xlabel('Category')
plt.ylabel('Amount (CAD)')
plt.title('Benefit Cash Inflows')
plt.xticks(index + bar_width / 2, df_benefit["Category"])
plt.legend()

# Annotate bars with actual numbers
for bar, value in zip(bar1, df_benefit["Shore Power"]):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{value:,.2f}', ha='center', va='bottom')

for bar, value in zip(bar2, df_benefit["Solar + Shore Power"]):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{value:,.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig("Benefit_Cash_Inflows.pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data Preparation
initial_costs_data = {
    "Type": ["Shore Power", "Solar + Shore Power"],
    "Initial Costs (CAD)": [0, 53336.29]
}

annual_operating_costs_data = {
    "Type": ["Shore Power", "Solar + Shore Power"],
    "Annual Operating Costs (CAD)": [28440.80, 13216.65]
}

payback_data = {
    "Type": ["Without 30% ITC", "With 30% ITC"],
    "Payback Period (Years)": [3.5, 2.4]
}

# DataFrames
df_initial_costs = pd.DataFrame(initial_costs_data)
df_annual_operating_costs = pd.DataFrame(annual_operating_costs_data)
df_payback = pd.DataFrame(payback_data)

# Additional Analysis: Monthly, Quarterly, and Yearly Savings with and without ITC

# Monthly Savings
monthly_savings_without_itc = 28440.80 / 12
monthly_savings_with_itc = 13216.65 / 12

# Quarterly Savings
quarterly_savings_without_itc = 28440.80 / 4
quarterly_savings_with_itc = 13216.65 / 4

# Yearly Savings
yearly_savings_without_itc = 28440.80
yearly_savings_with_itc = 13216.65

# Data Preparation for Savings Comparison
savings_data = {
    "Period": ["Monthly", "Quarterly", "Yearly"],
    "Without ITC (CAD)": [monthly_savings_without_itc, quarterly_savings_without_itc, yearly_savings_without_itc],
    "With ITC (CAD)": [monthly_savings_with_itc, quarterly_savings_with_itc, yearly_savings_with_itc]
}

df_savings = pd.DataFrame(savings_data)

# Plotting Savings Comparison
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35
bars1 = ax.bar(df_savings["Period"], df_savings["Without ITC (CAD)"], width, label='Without ITC')
bars2 = ax.bar(df_savings["Period"], df_savings["With ITC (CAD)"], width, bottom=df_savings["Without ITC (CAD)"], label='With ITC')
ax.set_title("Savings Comparison")
ax.set_ylabel("Savings (CAD)")
ax.legend()

# Annotate bars with actual numbers
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,.2f}', ha='center', va='bottom')

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + bars1[0].get_height(), f'{height:,.2f}', ha='center', va='bottom')

plt.savefig("Savings_Comparison2.pdf")
plt.show()
```
